```python
base_path = "/Users/tinazhang/Downloads/"

# national data
years = list(range(2001, 2024))  # From 2001 to 2023
msa_data = {}

# Loop through each year and read the corresponding Excel file
for year in years:
    file_name = f"MSA_M{year}_dl.csv"  # Construct the file name
    file_path = base_path + file_name  # Construct the full file path

    # Read the Excel file into a DataFrame
    df = pd.read_csv(file_path)
    try:
        data_filtered = df[['AREA', 'AREA_TITLE', 'OCC_CODE', 'TOT_EMP', 'A_MEAN', 'A_MEDIAN', 'ANNUAL', 'H_MEDIAN']]
    except: # if lowercase
        try:
            try:
                data_filtered = df[['area', 'area_name', 'occ_code', 'tot_emp', 'a_mean', 'a_median', 'annual', 'h_median']]
            except:
                data_filtered = df[['AREA', 'AREA_NAME', 'OCC_CODE', 'TOT_EMP', 'A_MEAN', 'A_MEDIAN', 'ANNUAL', 'H_MEDIAN']]
        except:
            data_filtered = df[['area', 'area_title', 'occ_code', 'tot_emp', 'a_mean', 'a_median', 'annual', 'h_median']]
            

    # Rename columns
    data_filtered.columns = ['AREA', "NAME", "OCC_CODE", "TOT_EMP", "A_MEAN", "A_MEDIAN", 'ANNUAL', "H_MEDIAN"]

    msa_data[year] = data_filtered
```
